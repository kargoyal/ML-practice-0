```python
def lo(x):
    for x in range(-6,4):
        y=1/(1+ny.exp(-x))
        
        ml.scatter(x,y,color="g",marker="D",s=79)
        ml.text(x,y,y,color="r")
        ml.title("Logit Values")
        ml.xlabel("x values")
        ml.ylabel("y=logit value")
```
